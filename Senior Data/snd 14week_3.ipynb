#1. 트리(Tree)의 앙상블(Ensemble)

##1-1. **앙상블 학습**

앙상블(Ensemble Learning)은 여러 개의 분류기를 생성하고, 그 예측을 결합함으로써 보다 정확한 예측을 구현하는 기법임 

앙상블 학습을 일반적으로 **보팅(Voting)**, **배깅(Bagging)**, **부스팅(Boosting)**의 세 가지 유형으로 구분됨

###1-1-1 **보팅(Voting)**

- 보팅은 여러개의 분류기가 투표를 통해 최족 예측결과를 결정하는 방식으로

- 서로 다른 알고리즘을 여러개 결합하여 사용한다.

- 보팅방식은 하드보팅과 소프트보팅으로 구분한다.

  **하드보팅(Hard Voting)** 
  : 다수의 분류기가 예측한 결과값을 최종 결과로 선정

  **소프트보팅(Soft Voting)** 
  : 모든 분류기가 예측한 레이블 값의 결정 확률 평균을 구한 뒤
    가장 확률이 높은 레이블의 값을 최종 결과로 선정

(참고) 서포트 벡터 머신(SVM, Support Vector Machine) : 주어진 데이터가 어느 카테고리에 속할지 판단하는 이진 선형 분류 모델

### 1-1-2. **배깅(Bagging)**

- **데이터 샘플링(Bootstrap)**을 통해 모델을 학습시키고 결과를 집계(Aggregating)하는 방법

- 모두 같은 유형의 알고리즘 기반의 분류기를 사용

- 데이터 분할 시 중복을 허용

- 방법

  **Categorical Data** : 다수결 투표 방식으로 결과 집계

  **Continuous Data** : 평균값 집계

- 과적합(Overfitting) 방지에 효과적

- 대표적인 배깅 방식 : **랜덤 포레스트 알고리즘**

image.png

(참고) **부스스트랩**은 데이터 세트에서 중복을 허용하여 데이터를 샘플링하는 방식

예) 비복원추출과 같이 1000개 샘플에서 하나의 샘플을 뽑고 샘플을 집어넣은 뒤에 다시 뽑는 경우 등
    

#### (참고) 랜덤포레스트

랜덤 포레스트는 아래 그림과 같이 **배깅(Bagging)** 방식으로 이루어진다. 

랜덤 포레스트 알고리즘은 대표적은 머신러닝 알고리즘으로 강력한 성능을 가지고 있는 것으로 알려져 있음 


1) 결정트리(Decision Tree) 분류기를 여러 개를 훈련하고,

2) 여러 개의 훈련한 결과를 가지고 예측함 

3) 그 과정에서 각각의 결정트리를 훈련시킬 때 **배깅(Bagging)**을 이용

3-1) 여기서 배깅은 **전체훈련 세트에서 중복을 허용하여 샘플링한 데이터 세트**를

3-2) **개별 결정트리의 훈련세트로 훈련**하는 방식임

4) 앙상블 방식을 사용하기 때문에 몇몇의 결정트리가 과대적합, 과소적합하더라도 경우에도 오차가 무시되는 경향이 있음

### **1-1-3**. **부스팅(Boosting)**

- **여러 개의 분류기가 순차적으로 학습을 수행**

- 이전 분류기가 예측이 틀린 데이터에 대해서 올바르게 예측할 수 있도록 다음 

- 분류기에게 가중치(weight)를 부여하면서 학습과 예측을 진행

- 계속하여 분류기에게 가중치를 부스팅하며 학습을 진행하기에 부스팅 방식이라 함

- 예측 성능이 뛰어나 앙상블 학습을 주도

- 대표적인 부스팅 모듈 – **XGBoost, LightGBM**

- 보통 부스팅 방식은 배깅에 비해 성능이 좋지만, 속도가 느리고 과적합이 발생할 
  가능성이 존재하므로 상황에 따라 적절하게 사용해야 함.

##1-2. **랜덤 포레스트 알고리즘 실습** 

image.png

- 랜덤 포레스트는 위 그림과 같이 여러 알고리즘은 결정트리(Decision Tree)들로 이루어 지기 때문에 

- 각각의 결정트리를 학습하는 과정이 필요함

  : 일반적인 결정트리 학습 -> **최적의 특성(노드 분할의 기준)을 찾고, 해당 특성을 잘 나눌 수 있는 임계점**을 찾는다.

  : **랜덤포레스트 방식 ->  **무작위로 선택한 후보들 중에서 최적의 특성**을 찾는다.


이렇게 무작위로 선택한 특성 후보들 중에서 최적의 특성을 사용한다면

1) 항상 최적의 특성을 찾는 방식 보다 빠르다.

2) 분산이 작아져서 결정경계를 쉽게 만들 수 있음 

**[교재 예시]**

1000개의 가방에서 100개 씩의 샘플을 뽑을 때 이를 부트스트랩 방식으로 할 경우

1) 1개를 뽑고, 뽑았던 1개를 다시 가방에 넣는다. 

2) 위 1)과 같은 방식으로 계속해서 100개의 가방을 뽑는다.

3) 일반적으로 부트스트랩 샘플을 훈련세트와 같은 크기로 만듬 
   
   훈련 세트가 만약에 1000개라면, 1000개의 샘플을 뽑는다. (훈련 세트와 크기는 동일)




image.png

이후 각 노드를 분할 할 때

1) 전체 특성 중에서 일부 특성을 무작위로 고른 다음

2) 최선의 분할을 찾는다.

3) **RandomForestClassfier는 전체 특성계수의 제곱근 만큼의 특성을 선택** 함 

  (예) 4개의 특성이 있다면 노드마다 2개를 랜덤하게 선택해서 사용
  
  (예외) RandomForestRegressor (전체특성 사용)

image.png

```python
#데이터를 훈련용, 테스트용으로 분할해서 사용 

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

wine = pd.read_csv('https://bit.ly/wine_csv_data')

data = wine[['alcohol', 'sugar', 'pH']].to_numpy()
target = wine['class'].to_numpy()

train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
#cross_validate로 교차검증수행
from sklearn.model_selection import cross_validate

#랜덤포레스트 사용
from sklearn.ensemble import RandomForestClassifier

#랜덤forest를 사용하는데 n_jobs=-1로 모든 cpu코어를 사용
rf = RandomForestClassifier(n_jobs=-1, random_state=42)

#return_Train_scores의 의미 : 검증세트 점수, 훈련세트에 대한 점수도 보여줌
scores = cross_validate(rf, train_input, train_target, return_train_score=True, n_jobs=-1)

#훈련세트에 대한 점수, 테스트 세트에 대한 점수를 봄 
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

훈련세트 점수 : 99.7

테스트 세트 점수 : 89.05 (과적합)

```python
#특성 중요도를 보여주고 이를 비교 
#알콜도수, 당도, ph가 중요했음

rf.fit(train_input, train_target)
print(rf.feature_importances_)
```

당도의 중요도는 떨어졌고, 알콜 도수와 ph 중요도가 상승했음

(참고) 앞에서 구한 결정 트리의 중요도는 0.123456...0.8686... 0.00791...이있음 

```python
#자체적으로 모델을 평가하는 점수를 계산할 수 있음

# 어떻게 -> 중복을 허용하는 샘플을 만들다보면, 샘플에 포함되지 않고 남는 샘플 있음
# 이 샘플(OOB) 이용해서 검정세트의 역할을 맡긴다.

#obb_score 라는 옵션을 사용
rf = RandomForestClassifier(oob_score=True, n_jobs=-1, random_state=42)

#obb_scores 라는 옵션을 사용
rf.fit(train_input, train_target)
print(rf.oob_score_)
```

#1-3. 엑스트라트리(Extra Tree)


랜덤 포레스트 방식과 동일하지만 부트스트랩 tree를 사용하지는 않는다.

- 각 결정트리를 만들 때 **전체훈련세트** 를 사용함

- 노드를 분할할 때 가장 좋은 분할을 고르는 것이 아니라 무작위로 분할 





```python
#엑스트라트리를 사용
from sklearn.ensemble import ExtraTreesClassifier

et = ExtraTreesClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(et, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
#특성에 대한 중요도를 출력 
et.fit(train_input, train_target)
print(et.feature_importances_)
```

(결과 해석) 랜덤 포레스트 트리와 마찬가지로 당도의 의존성이 감소

##1-4. **그레이디언트 부스팅**

깊이가 얕은 결정트리를 이용해서 이전 트리의 오차를 보완하는 방법 

```python
#그레디언트 트리를 사용함 
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

훈련용 세트 정확도 : 0.888

테스트용 세트 정확도 : 0.872 (과대적합을 상당부분 개선)

```python
#트리개수를 500개로 할 경우에도 과대적합이 나타나지 않음
#learning_rate의 기본 값 : 0.2 

gb = GradientBoostingClassifier(n_estimators=500, learning_rate=0.2, random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
#앞의 모형과 마찬가지로 당도에 집중하는 경향을 보임
gb.fit(train_input, train_target)
print(gb.feature_importances_)
```

##1-5. **히스토그램 기반 그래디언트 부스팅**

(이후 내용부터는 다양한 머신러닝 알고리즘을 적용해서 정확도를 높이는 작업)

```python
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier

hgb = HistGradientBoostingClassifier(random_state=42)
scores = cross_validate(hgb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
from sklearn.inspection import permutation_importance

hgb.fit(train_input, train_target)
result = permutation_importance(hgb, train_input, train_target, n_repeats=10,
                                random_state=42, n_jobs=-1)
print(result.importances_mean)
```

```python
result = permutation_importance(hgb, test_input, test_target, n_repeats=10,
                                random_state=42, n_jobs=-1)
print(result.importances_mean)
```

```python
hgb.score(test_input, test_target)
```
